```python
import numpy as np
import matplotlib.pyplot as plt
```

$z_0 = 0i$

$z_{n+1} = z_n^2 + c$

$|z| \le 2$

$|z| = \sqrt{Re(z) + Im(z)} $

```python
def mandelbrot(x, y, max_iters):
    z = complex(0,0)
    c = complex(x, y)
    for i in range(max_iters):
        z = z * z + c 
        if abs(z) > 2:
            return i
    return max_iters
```

```python


def draw(image, x_points, y_points, max_iters):
    for ix, x in enumerate(x_points):
        for iy, y in enumerate(y_points):
            image[iy, ix] = mandelbrot(x, y, max_iters)
```

$ f(x,y) = xy$

```python
width = 1000
height = 1000

x_bounds = (-2,1)
y_bounds = (-1.5,1.5)

x_points = np.linspace(x_bounds[0], x_bounds[1], width)
y_points = np.linspace(y_bounds[0], y_bounds[1], height)

image = np.zeros((height, width))
MAX_ITERS=500
draw(image, x_points, y_points, max_iters=MAX_ITERS)

plt.imshow(image, cmap='plasma', interpolation='none')
plt.colorbar()
plt.show()
```
