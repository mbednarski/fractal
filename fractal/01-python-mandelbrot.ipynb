# Mandelbrot set

The algorithm is following.

For each pixel on a complex plane we check if it belongs to the set by:

$z_0 = 0i$

For $n \to +\inf$

$z_{n+1} = z_n^2 + c$

if $|z| \le 2$ then point belongs to the set

$|z|$ is defined as follows:

$|z| = \sqrt{Re(z)^2 + Im(z)^2} $

## Function that determines if specific point belongs to the set

```python
def mandelbrot(x, y, max_iters):
    z = complex(0,0)
    c = complex(x, y)
    for i in range(max_iters):
        z = z * z + c 
        if abs(z) > 2:
            return i
    return max_iters
```

## Function that draws fractal on specific plane

```python
def draw(image, x_points, y_points, max_iters):
    for ix, x in enumerate(x_points):
        for iy, y in enumerate(y_points):
            image[iy, ix] = mandelbrot(x, y, max_iters)
```

## Generating the fractal

```python
# image resolution
width = 1000 
height = 1000

# complex plane bounds - keep same x/y ration as in resolution to avoid stretched image
x_bounds = (-2,1)
y_bounds = (-1.5,1.5)

# number of iterations
MAX_ITERS=50

x_points = np.linspace(x_bounds[0], x_bounds[1], width)
y_points = np.linspace(y_bounds[0], y_bounds[1], height)

# create blank image
image = np.zeros((height, width))

# draw it
draw(image, x_points, y_points, max_iters=MAX_ITERS)

# and see!
plt.imshow(image, cmap='plasma', interpolation='none') # do not interpolate in order to see true values
plt.colorbar()
plt.show()
```
